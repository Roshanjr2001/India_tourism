```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from pandas.plotting import register_matplotlib_converters

# Register converters for time series plots
register_matplotlib_converters()

# Original data

# Convert data into a DataFrame
df = pd.read_excel('/home/agira/Downloads/Indian Tourism (1).xlsx', skiprows=1)
cols_to_keep = ['Months', 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
df = df[cols_to_keep]
df
```

```python
# Convert data to a time series format
df = df.melt(id_vars=["Months"], var_name="Year", value_name="Arrivals")

df
```

```python
# Convert Year-Month to a datetime index
df["Date"] = pd.to_datetime(df["Year"].astype(str) + "-" + df["Months"].str.strip(), format="%Y-%B")

# Set Date as index and remove old columns
df = df.drop(columns=["Months", "Year"]).set_index("Date")

print(df)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x=df.index, y="Arrivals", label="Monthly Arrivals")
plt.xlabel("Year")
plt.ylabel("Number of Arrivals")
plt.title("Monthly Arrivals Over Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

```python
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(df['Arrivals'])
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(df['Arrivals'])
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf

plot_pacf(df['Arrivals'])
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose


decomposition=seasonal_decompose(df['Arrivals'],model='Multiplative',period=12)
decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller




result = adfuller(df.Arrivals, autolag='AIC')
p_value = result[1]
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {p_value}')
print('Stationary' if p_value < 0.05 else 'Non-Stationary')
```

```python
df['Arrivals_MA'] = df['Arrivals'].rolling(window=12).mean()
df['Arrivals_detrended'] = df['Arrivals'] - df['Arrivals_MA']
df['Arrivals_detrended_diff'] = df['Arrivals_detrended'].diff()

dftest_detrended = adfuller(df['Arrivals_detrended_diff'].dropna(), autolag='AIC')
print("ADF Statistic (detrended + diff):", dftest_detrended[0])
print("P-Value (detrended + diff):", dftest_detrended[1])
print('Stationary' if dftest_detrended[1] < 0.05 else 'Non-Stationary')
```

```python

from statsmodels.tsa.statespace.sarimax import SARIMAX

# Fit the SARIMAX model
sarima = SARIMAX(df['Arrivals'], order=(1,1,1), seasonal_order=(0,1,0,12))
sarima_fit = sarima.fit()
future_dates = pd.date_range(start='2020-01-01', periods=48, freq='ME')  # 48 months (4 years)

# Forecast future values
forecast_values = sarima_fit.forecast(steps=48)  # Predict for 48 months

# Convert forecast to DataFrame
forecast_df = pd.DataFrame({'Date': future_dates, 'Forecast': forecast_values})
forecast_df.set_index('Date', inplace=True)
df_combined = pd.concat([df, forecast_df], axis=1)

# Save to CSV
df_combined.to_csv('combined_data.csv', index=True)

print("Combined data saved successfully!")
# Display forecasted values
print(forecast_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot historical data
plt.figure(figsize=(12,6))
plt.plot(df.index, df['Arrivals'], label='Actual Arrivals', color='blue')

# Plot forecasted values
plt.plot(forecast_df.index, forecast_df['Forecast'], label='Forecast (2020-2023)', color='red', linestyle='dashed')

# Formatting the plot
plt.xlabel('Year')
plt.ylabel('Arrivals')
plt.title('SARIMA Forecast for Arrivals (2020-2023)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
